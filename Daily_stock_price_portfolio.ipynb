```python
adjclose = []

#Apple stock price 
yahoo_financials_APPL = YahooFinancials('AAPL')
data_appl = yahoo_financials_APPL.get_historical_price_data(start_date='2020-01-01', 
                                                  end_date='2022-2-2', 
                                                  time_interval='monthly')
aapl_df = round(pd.DataFrame(data_appl['AAPL']['prices']),2)
aapl_df = aapl_df.drop('date', axis=1).set_index('formatted_date')
adjclose.append(aapl_df["adjclose"])

#aapl_df["adjclose"].plot(title="Daily stock price")


#Facebook stock price 
yahoo_financials_FB = YahooFinancials('FB')
data_fb = yahoo_financials_FB.get_historical_price_data(start_date='2020-01-01', 
                                                  end_date='2022-2-2', 
                                                  time_interval='monthly')
fb_df = round(pd.DataFrame(data_fb['FB']['prices']),2)
fb_df = fb_df.drop('date', axis=1).set_index('formatted_date')
adjclose.append(fb_df["adjclose"])

#Amazon stock price 
yahoo_financials_AMZN = YahooFinancials('AMZN')
data_amzn = yahoo_financials_AMZN.get_historical_price_data(start_date='2020-01-01', 
                                                  end_date='2022-2-2', 
                                                  time_interval='monthly')
amzn_df = round(pd.DataFrame(data_amzn['AMZN']['prices']),2)
amzn_df = amzn_df.drop('date', axis=1).set_index('formatted_date')
adjclose.append(amzn_df["adjclose"])

#Microsoft stock price 
yahoo_financials_MSFT = YahooFinancials('MSFT')
data_msft = yahoo_financials_MSFT.get_historical_price_data(start_date='2020-01-01', 
                                                  end_date='2022-2-2', 
                                                  time_interval='monthly')
msft_df = round(pd.DataFrame(data_msft['MSFT']['prices']),2)
msft_df = msft_df.drop('date', axis=1).set_index('formatted_date')
adjclose.append(msft_df["adjclose"])

#Google stock price 
yahoo_financials_GOOGL = YahooFinancials('GOOGL')
data_googl = yahoo_financials_GOOGL.get_historical_price_data(start_date='2020-01-01', 
                                                  end_date='2022-2-2', 
                                                  time_interval='monthly')
googl_df = round(pd.DataFrame(data_googl['GOOGL']['prices']),2)
googl_df = googl_df.drop('date', axis=1).set_index('formatted_date')
adjclose.append(googl_df["adjclose"])

#Tesla stock price 
yahoo_financials_TSLA = YahooFinancials('TSLA')
data_tsla = yahoo_financials_TSLA.get_historical_price_data(start_date='2020-01-01', 
                                                  end_date='2022-2-2', 
                                                  time_interval='monthly')
tsla_df = round(pd.DataFrame(data_tsla['TSLA']['prices']),2)
tsla_df = tsla_df.drop('date', axis=1).set_index('formatted_date')
adjclose.append(tsla_df["adjclose"])

#NVIDIA stock price 
yahoo_financials_NVDA = YahooFinancials('NVDA')
data_nvda = yahoo_financials_NVDA.get_historical_price_data(start_date='2020-01-01', 
                                                  end_date='2022-2-2', 
                                                  time_interval='monthly')
nvda_df = round(pd.DataFrame(data_nvda['NVDA']['prices']),2)
nvda_df = nvda_df.drop('date', axis=1).set_index('formatted_date')
adjclose.append(nvda_df["adjclose"])

#JP Morgan stock price 
yahoo_financials_JPM = YahooFinancials('JPM')
data_jpm = yahoo_financials_JPM.get_historical_price_data(start_date='2020-01-01', 
                                                  end_date='2022-2-2', 
                                                  time_interval='monthly')
jpm_df = round(pd.DataFrame(data_jpm['JPM']['prices']),2)
jpm_df = jpm_df.drop('date', axis=1).set_index('formatted_date')
adjclose.append(jpm_df["adjclose"])

#transform to a data frame, and then transpose it, Every Column will be a different ticker

df_adjclose = pd.DataFrame(adjclose)
T_df_adjclose = df_adjclose.T

T_df_adjclose.columns = ["AAPL", "FB","AMZN","MSFT","GOOGL","TSLA","NVDA","JPM"]
plt.figure(figsize=(100,100))
T_df_adjclose.plot(title = "Monthly Stock Price")
plt.xlabel("Date") 
plt.ylabel("Price")
plt.show()
T_df_adjclose.head()
```
